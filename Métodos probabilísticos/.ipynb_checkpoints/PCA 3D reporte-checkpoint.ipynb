# PCA 3D imágenes

Primero importamos las bibliotecas necesarias

```python
from PIL import Image #oara lee coordenadas rgb
import numpy as np # para operaciones matriciales
import matplotlib.pyplot as plt # para los plots
from IPython.display import Image as IM #para mostrar la imagen de la que se hará el análisis
```

## Esta función toma la imagen y crea un array con las coordenadas (r,g,b)

```python
def convertir_a_rgb(ruta_imagen):
    imagen =  Image.open(ruta_imagen)
    pixeles = list(imagen.getdata())
    print(len(pixeles))
    r_c = [pixel[0] for pixel in pixeles]
    g_c = [pixel[1] for pixel in pixeles]
    b_c = [pixel[2] for pixel in pixeles]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(r_c, g_c, b_c, c='green', marker='x', alpha=0.8, s=1)
    ax.set_xlabel('R')
    ax.set_ylabel('G')
    ax.set_zlabel('B')
    plt.show()
    
    return np.array((r_c, g_c, b_c))
```

## Estas funciones se utilizan para calcular la matriz de covarianza, de correlación y los eigenvectores con sus respectivos eigenvalores

```python
def promedio(array):
    promedio = sum(array) / len(array)
    return promedio

def cov(X,Y):
    mean_X = promedio(X)
    mean_Y = promedio(Y)
    n = len(X)
    covariance = sum((X[i] - mean_X) * (Y[i] - mean_Y) for i in range(n)) / (n - 1)
    return covariance

def matriz_covarianza(array):
    n = len(array)
    matriz_covarianza = [[0 for _ in range(n)] for _ in range(n)]
    for i in range(len(array)):
        for j in range(len(array)):
            matriz_covarianza[i][j]=cov(array[i],array[j])
    print("la matriz de covarianza es")
    for fila in matriz_covarianza:
        print(fila)
    return np.array(matriz_covarianza)

def matriz_correlacion(array):
    n = len(array)
    matriz_cov = matriz_covarianza(array)
    matriz_correlacion = np.ones((n, n))
    
    for i in range(n):
        for j in range(n):
            if i!=j:
                std_i = np.std(array[i])
                std_j = np.std(array[j])
                matriz_correlacion[i][j] = matriz_cov[i][j] / (std_i * std_j)
    
    print("La matriz de correlación es:")
    for fila in matriz_correlacion:
        print(fila)
    
    return matriz_correlacion

def eigen(matriz):    
    eigenvalores, eigenvectores = np.linalg.eig(matriz.T)
    print("Eigenvalores:")
    print(eigenvalores)
    print("--------------------------------------")
    print("\nEigenvectores:")
    print(eigenvectores)
    suma_total_eigenvalores = np.sum(eigenvalores)
    proporcion_varianza = eigenvalores / suma_total_eigenvalores
    eigenvectores = eigenvectores.T
    print("--------------------------------------")
    for eigenvalor, prop in zip(eigenvectores, proporcion_varianza):
        print(f"\n Eigenvector {eigenvalor}| proporcion {prop:.4f}")
    return [eigenvalores, eigenvectores]

    
```

## Finalmente estas funciones realizan los plots necesarios

```python
def centroide_plot(p):
    vector_prom = [promedio(p[0]), promedio(p[1]), promedio(p[2])]
    r,g,b=vector_prom
    origen = np.zeros(3)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.quiver(*origen, *[r,g,b],color="green", linestyle='-', linewidth=1, label='Vector promedio')
    ax.scatter(r,g,b, s=5, c='red', label="Centroide")
    ax.set_xlabel('R')
    ax.set_ylabel('G')
    ax.set_zlabel('B')
    ax.legend()
    plt.show()
    
```

```python
def ejes_principales(eigenvectores):
    origen = np.zeros(3)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for eigenvector, color in zip(eigenvectores, ['#F6A21E', '#E55B13', '#7A871E']):
        ax.quiver(*origen, *eigenvector, color=color)
    ax.set_xlabel('R')
    ax.set_ylabel('G')
    ax.set_zlabel('B')
    plt.show()
    
```

```python
def plot_completo(eigenvectores, eigenvalores, array):
    x=array[0]
    y=array[1]
    z=array[2]
    vector_prom = [promedio(x), promedio(y), promedio(z)]
    r,g,b=vector_prom
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    origen = np.zeros(3)
    for eigenvector, color, eigenvalue in zip(eigenvectores, ['#581845', '#C70039', '#FF5733'], eigenvalores):
        eigenvector_shifted = eigenvector * eigenvalue + [r, g, b]  # Sumar el vector promedio
        ax.quiver(r, g, b, *eigenvector_shifted, color=color, label=f'Eigenvector {eigenvector}')

    ax.scatter(r,g,b, s=100, c='red')
    ax.scatter(x, y, z, c='#48DB6E', marker='x', alpha=0.5, s=1, label='Datos originales')

    ax.set_xlabel('R')
    ax.set_ylabel('G')
    ax.set_zlabel('B')
    plt.show()
```

# Primera imagen

Para este primer ejercicio elegí una imagen que tuviera predominantemente el color azul con el propósito de lograr ver valores rojo y verde pequeños

```python
ruta_imagen = 'azul.jpg'
IM(filename=ruta_imagen)
```

Primero convertimos a coordenadas rgb y hacemos un plot de los datos

```python
pixeles=convertir_a_rgb('azul.jpg')
```

Podemos ver que, en efecto, la mayoria de los puntos se acumulan en el plano $R=0$ pues es el color que menos aparece en la imagen

```python
matriz_cov =matriz_covarianza(pixeles)
```

En este caso, los valores son aproximadamente 29.11 para la varianza de R, 4329.93 para la varianza de G y 10523.72 para la varianza de B. Esto indica cuánto varían individualmente los canales R, G y B en la imagen. Naturalmente hay mayor varianza en el color azul casi el doble que el del color verd

```python
matriz_cor = matriz_correlacion(pixeles)
```

Al calcular la matriz de correlación podemos ver que naturalmente no hay correlación entre el canal R y el resto mientras que la relación entre G y B es casi uno, esto puede interpretarse como que el color azul predominante es una combinación de éstos canales.

```python
eigen_result = eigen(matriz_cov)
```

```python
centroide_plot(pixeles)
```

```python
ejes_principales(eigen_result[1])
```

```python
plot_completo(eigen_result[1],eigen_result[0],pixeles)
```

Finalmente viendo los eigenvectores notamos que la mayoría de los puntos pueden expresarse como aquellos que son una combinación de verde con azul y una coordenada roja muy pequeña y que aquellos colores que salen de esta tendencia son en su mayoría de colores azules más intentos.

En conlcución:
Los canales G y B tienen una fuerte relación lineal positiva, lo que significa que tienden a aumentar juntos en la imagen.
Los canales R y G tienen una relación lineal negativa más débil, lo que significa que tienden a moverse en direcciones opuestas, pero la relación es menos fuerte que en el caso de G y B.
Los canales R y B también tienen una relación lineal positiva, pero es más débil que la de G y B.

# Tercera imagen

Para esta tercera imagen utilicé una imagen generada de la siguiente manera: se genera un grid de 250 x 250 pixeles, cada pixel tiene coordenadas R,G,B. El color R se asigna de manera aleatoria, el valor G se asigna de manera aleatoria (entre 0 y 126) independiente de R y finalmente el valor de B se asigna multiplicando el valor de G multiplicado por un número aleatorio entre 1 y 2 para no exceder el valor de 256 permitido. Veamos cómo luce la imagen

```python
ruta_imagen = 'imagen_generada.png'
IM(filename=ruta_imagen)
```

```python
pixeles3 = convertir_a_rgb('imagen_generada.png')
```

La imagen muestra 3 planos: el primero del resultado de crear un "espacio" de dimensión 2 pues en la generación de los puntos sólo utilizamos dos coordenads, y los otros dos los resultados de multiplicar por 1 o por 2 los elementos que podrían estar en el primer plano.

```python
matriz_cov3 =matriz_covarianza(pixeles3)
```

Los elementos en la diagonal principal de la matriz representan las varianzas de las variables RGB individuales. En este caso, las varianzas son muy altas en comparación con las matrices anteriores. Esto indica que las variables R, G y B tienen una variabilidad individual extremadamente alta en la imagen.

Los elementos fuera de la diagonal principal representan las covarianzas entre las variables. En este caso, la covarianza entre R y G (0.718) y la covarianza entre G y B (1342.241) son positivas, mientras que la covarianza entre R y B (-21.758) es negativa. Esto sugiere que en la imagen, R y G tienden a aumentar juntas, G y B también tienden a aumentar juntas, pero R y B tienden a moverse en direcciones opuestas.

```python
matriz_cor3 = matriz_correlacion(pixeles3)
```

La correlación entre R y G es muy cercana a cero (0.0003), lo que indica una correlación muy débil o prácticamente inexistente entre estos dos canales.
La correlación entre G y B es significativamente positiva (0.526), lo que sugiere una correlación moderada entre estos dos canales.
La correlación entre R y B es negativa (-0.0042), pero también es muy cercana a cero, lo que sugiere una correlación muy débil o prácticamente inexistente entre estos dos canales.

```python
eigen_result3 = eigen(matriz_cov3)
```

```python
centroide_plot(pixeles3)
```

```python
ejes_principales(eigen_result3[1])
```

```python
plot_completo(eigen_result3[1],eigen_result3[0],pixeles3)
```

Los canales R y G tienen una correlación muy débil, lo que significa que no están relacionados linealmente en la imagen.
Los canales G y B tienen una correlación moderada, lo que indica que tienden a moverse juntos, pero la relación no es extremadamente fuerte.
Los canales R y B tienen una correlación muy débil o prácticamente inexistente.
