# 1.Grafica en 3D los datos RGB de una imagen. Un pixel tendría (x,y,R,G,B)

```python
imagen = Image.open('azul.jpg')  
```

```python
pixeles = list(imagen.getdata())
print(len(pixeles))
```

```python
# pixeles=pixeles[:400000]
# print(len(pixeles))
print(type(pixeles))
```

```python
print(pixeles[0])
```

```python
r_c = [pixel[0] for pixel in pixeles]
g_c = [pixel[1] for pixel in pixeles]
b_c = [pixel[2] for pixel in pixeles]
print(len(r_c))
```

```python
rojos = [r for r in r_c if r==0]
green = [g for g in g_c if g==0]
blue = [b for b in b_c if b==0]
print(len(rojos))
print(len(green))
print(len(blue))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r_c, g_c, b_c, c='green', marker='x', alpha=0.8, s=1)
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')
plt.show()
```

## 2. Calcula la matriz de covarianza 

```python
def promedio(array):
    promedio = sum(array) / len(array)
    return promedio

def cov(X,Y):
    mean_X = promedio(X)
    mean_Y = promedio(Y)
    n = len(X)
    covariance = sum((X[i] - mean_X) * (Y[i] - mean_Y) for i in range(n)) / (n - 1)
    return covariance
```

```python
matriz_covarianza=[
    [cov(r_c,r_c), cov(r_c,g_c), cov(r_c,b_c) ],
    [cov(r_c,g_c), cov(g_c,g_c), cov(b_c,g_c) ],
    [cov(r_c,b_c), cov(b_c,g_c), cov(b_c,b_c)]
]
```

```python
#matriz_covarianza = np.cov(pixeles)
for fila in matriz_covarianza:
    print(fila)
```

## 3. Calcula la matriz de correlación

```python
matriz_correlacion = np.corrcoef(pixeles, rowvar=False)
```

```python
print(matriz_correlacion)
```

## 4. Calcula los 3 eigenvalores y 3 eigenvectores

```python
eigenvalores, eigenvectores = np.linalg.eig(matriz_covarianza)
```

```python
print("Eigenvalores:")
print(eigenvalores)

print("\nEigenvectores:")
print(eigenvectores)
```

## 4. Imprimir qué proporción de la varianza de los datos explica cada eigenvalor

```python
suma_total_eigenvalores = np.sum(eigenvalores)
proporcion_varianza = eigenvalores / suma_total_eigenvalores
#print(sum(proporcion_varianza_explicada))

for i, prop in enumerate(proporcion_varianza):
    print(f"Eigenvector {i + 1}: {prop:.4f}")
    
for eigenvalor, prop in zip(eigenvectores, proporcion_varianza):
        print(f"Eigenvector {eigenvalor}| proporcion: {prop:.4f}")
```

## 5. Graficar el centro o valor medio

```python
vector_prom = [promedio(r_c), promedio(g_c), promedio(b_c)]
print(vector_prom)
r,g,b=vector_prom
```

```python
origen = np.zeros(3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(*origen, *[r,g,b],color="green", linestyle='-', linewidth=1, label='Vector promedio')
ax.scatter(r,g,b, s=5, c='red', label="Centroide")
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')
ax.legend()
plt.show()
```

## 6. Graficar los ejes principales

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for eigenvector, color in zip(eigenvectores, ['#F6A21E', '#E55B13', '#7A871E']):
    ax.quiver(*origen, *eigenvector, color=color)
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origen = np.zeros(3)
for eigenvector, color, eigenvalue in zip(eigenvectores, ['#581845', '#C70039', '#FF5733'], eigenvalores):
    eigenvector_shifted = eigenvector * eigenvalue + [r, g, b]  # Sumar el vector promedio
    ax.quiver(r, g, b, *eigenvector_shifted, color=color, label=f'Eigenvector {eigenvector}')
    
ax.scatter(r,g,b, s=10, c='red')
ax.scatter(r_c, g_c, b_c, c='#9AD0F1', marker='x', alpha=0.5, s=1, label='Datos originales')

ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')
plt.show()

```
